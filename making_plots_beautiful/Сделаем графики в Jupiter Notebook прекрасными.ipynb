Следовательно, мы бы рассматривали Matplotlib для построения графика. 
Matplotlib - это комплексная библиотека для создания статических, анимированных и интерактивных визуализаций на Python. 
matplotlib.pyplot - это набор функций командного стиля, который позволяет matplotlib работать как MATLAB. 
Функция pyplot может быть использована для создания рисунка, создания области построения на рисунке, построения некоторых линий в области построения, украшения графика метками и т.д. 
Визуализации могут быть быстро сгенерированы с помощью pyplot.

Давайте создадим некоторый код в Jupyter notebook, чтобы создать обычный дистрибутив. 
Для непосвященных нормальное распределение - это непрерывное распределение вероятностей для вещественной случайной величины. 
Его можно легко идентифицировать по колоколообразной кривой (функция плотности вероятности) и ее симметрии.

```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as stats
mu = 0
std = 1
x = np.linspace(start=-4, stop=4, num=100)
y = stats.norm.pdf(x, mu, std) 
plt.plot(x, y)
plt.show()
```

Код создает простой график нормального распределения со средним значением =0 и стандартным отклонением=1. 
Поскольку наша главная задача - сделать графики более красивыми, объяснение кода о математических аспектах будет довольно кратким. 
np.linspace() возвращает равномерно распределенные выборки (количество выборок равно num) за определенный интервал [start, stop]. 
scipy.stats.norm() возвращает нормальная непрерывная случайная величина. 
scipy.stats.norm.pdf() вычисляет PDF в любой точке для заданного значения среднего(mu) и стандартного отклонения(std).

График кажется слишком заурядным и пресным. Здесь нет ни ярлыков, ни названий, которые могли бы предоставить какую-то ценную информацию третьему лицу. Здесь нет сетки, позволяющей легко идентифицировать и сопоставлять значения. Размер фигуры тоже немного маловат, на мой вкус.

Давайте сделаем графики красивыми, используя различные функции pyplot.

### Добавление Линий Сетки

Сетки помогают легко идентифицировать и сопоставлять значения на графике. 
plt.grid() настраивает линии сетки. Он показывает сетку в настройках по умолчанию.

    plt.grid()

Простой код для создания фигуры выглядит следующим образом:

```python
import matplotlib.pyplot as plt
fig = plt.figure()
ax = plt.axes()
plt.grid()
plt.show()
```

plt.figure() создает новый рисунок. 
plt.axes() присваивает ось новому рисунку и делает ее текущей осью (множество осей). 
Рисунок Matplotlib можно рассматривать как единый контейнер, содержащий всю информацию об осях, графике, тексте и метках. 
Оси можно рассматривать как ограничивающую рамку с галочками и метками, которые будут содержать элементы графика визуализации. 
plt.show() отображает все фигуры и блокирует до тех пор, пока фигуры не будут закрыты.

Давайте поиграем с различными аспектами grid(). 
Начнём с цвета сетки.

```python
fig = plt.figure()
ax = plt.axes()
plt.grid(b=True, color='r')
```

*color* - это аргумент ключевого слова, который присваивает сетке цвет. 
Необязательный параметр *b* принимает логические значения (True или False). Если установлено значение False, сетки исчезнут. Однако, если присутствуют какие-либо аргументы ключевого слова (например, альфа, цвет, ширина строки и т.д.), То для *b* будет установлено значение True независимо от заданного значения b. Например, если приведенный выше фрагмент кода будет изменен:

```python
fig = plt.figure()
ax = plt.axes()
plt.grid(b=False, color = 'r')  
# plt.grid(b=False) 
```

...Результатом будет тот же график с сеткой красного цвета, что и показано выше.

Для *color* в качестве значений можно использовать любую из следующих строк:

Общие названия таких цветов, как красный, синий, коричневый, пурпурный и т.д.
####    color =’purple’
Цветовой шестнадцатеричный код #RRGGBB со значениями от 00 до FF. Здесь R =Красный, G = Зеленый, B = Синий
####    color = ‘#e3e3e3’
Кортеж RGB со значениями от 0 до 1.
####    color = (0.6, 0.9, 0.2)
Оттенки серого со значениями от 0.0 до 1.0. По мере увеличения значения черные линии сетки тускнеют.
####    color = ‘0.9’
Короткие цветовые коды для RGB и CMYK. Здесь r: красный, g: зеленый, b: синий, c: голубой, m: пурпурный, y: желтый, k: черный.
####    color=’c’
Имена цветов HTML. Вы можете найти больше названий цветов HTML здесь - https://www.w3schools.com/colors/colors_names.asp
####    color=’tomato’

```python
fig = plt.figure()
ax = plt.axes()
plt.grid(b=True, color='aqua', alpha=0.3, linestyle='-.', linewidth=2)
```

Свойства различных аргументов ключевого слова в коде указаны ниже:

*alpha* используется для изменения непрозрачности сетки. Нулевое значение для альфа приведет к исчезновению сетки.
*linestyle* или *ls* зададут стиль линий сетки. Подходящими значениями для linestyle являются '-', ' — ', '-.', ':', ', ' сплошная", "пунктирная", "пунктирная", "пунктирная точка" и т.д.
*linewidth* или *lw* определяет ширину линий сетки.

Примечание: Кривые или линии графика могут быть оформлены так же, как и сетка, с использованием тех же аргументов ключевого слова.

```python
import numpy as np
import matplotlib.pyplot as plt
fig = plt.figure()
ax = plt.axes()
x = np.linspace(0, 6.28, 100)
plt.plot(x, np.sin(x), color='Indigo', linestyle='--', linewidth=3)
plt.grid(b=True, color='aqua', alpha=0.3, linestyle='-', linewidth=1)
plt.show()
```

Аргументы *color* и *linestyle* можно объединить, чтобы получить аргумент, не связанный с ключевым словом. 
Например, 'c- -' эквивалентно установке color='cyan', linestyle='- -', как показано в приведенном ниже коде:

```python
import numpy as np
import matplotlib.pyplot as plt
fig = plt.figure()
ax = plt.axes()
x = np.linspace(0, 6.28, 100)
plt.plot(x, np.sin(x), 'c--', linewidth=3)
plt.grid(b=True, color='aqua', alpha=0.3, linestyle='-.', linewidth=2)
plt.show()
```

Это сочетание цвета и стиля линии не будет работать с grid().

### Основные и Второстепенные Линии сетки

Основные и второстепенные линии сетки можно отобразить, указав любое из трех значений параметра plt.grid(). 
Эти три значения являются основными, второстепенными и обоими. Они не требуют пояснений, как следует из их названия. Значение *major* отображает основные линии сетки, в то время как *minor* отображает второстепенные линии сетки. 
Значение *both* отображает как основные, так и второстепенные линии сетки.

```python
import numpy as np
import matplotlib.pyplot as plt
x = np.linspace(0, 7, 100)
line1, = plt.plot(x, np.sin(x), label='sin')
line2, = plt.plot(x, np.cos(x), label='cos')
plt.legend(handles=[line1, line2], loc='lower right')
#основные линии сетки
plt.grid(b=True, which='major', color='gray', alpha=0.6, linestyle='dashdot', lw=1.5)
#второстепенные линии сетки
plt.minorticks_on()
plt.grid(b=True, which='minor', color='beige', alpha=0.8, ls='-', lw=1)
plt.show()
```

plot.minorticks_on() отображает второстепенные отметки на осях. 
Недостатком является то, что это может снизить производительность.

Еще пример (токи в трехфазной сети):

```python
import numpy as np
import matplotlib.pyplot as plt
PI = 3.151592
x = np.linspace(0, 3.14*4, 100)
phase_A, = plt.plot(x, np.sin(x), label='фаза A')
phase_B, = plt.plot(x, np.sin(x+2*PI/3), label='фаза B')
phase_C, = plt.plot(x, np.sin(x+4*PI/3), label='фаза C')
plt.legend(handles=[phase_A, phase_B, phase_C], loc='lower right')
#основные линии сетки
plt.grid(b=True, which='major', color='gray', alpha=0.6, linestyle='dashdot', lw=1.5)
#второстепенные линии сетки
plt.minorticks_on()
plt.grid(b=True, which='minor', color='beige', alpha=0.8, ls='-', lw=1)
plt.show()
```

### Установка Предела осей

Matplotlib устанавливает ограничения осей по умолчанию, если ограничения осей специально не указаны. 
Вы можете установить ограничения с помощью plt.xlim() и plt.ylim для установки пределов x и y пределов соответственно.

```python
import numpy as np
import matplotlib.pyplot as plt
fig = plt.figure()
ax = plt.axes()
x = np.linspace(0, 5, 100)
plt.plot(x, np.sin(x), 'c--', linewidth=3)
# установлен предел по оси X - от -1 до +6
plt.xlim(-1, 6)
# установлен предел по оси Y - от -1.25 до +1.25
plt.ylim(-1.25, 1.25)
plt.grid(b=True, color='aqua', alpha=0.3, linestyle='-.', linewidth=2)
plt.show()
```

### Параметры меток

Мы видели, как настраивать сетки. 
Тем не менее, существует метод plot.tick_params(), который может изменять внешний вид меток, надписей меток и линий сетки в одной строке кода. Разве это не удобно и мощно?

```python
import numpy as np
import matplotlib.pyplot as plt
fig = plt.figure()
ax = plt.axes()
x = np.linspace(0, 5, 100)
plt.plot(x, np.sin(x), 'c--', linewidth=3)
plt.xlim(-1, 6)
plt.ylim(-1.25, 1.25)
plt.tick_params(axis='both', direction='out', length=6, width=2, labelcolor='b', colors='r', grid_color='gray', grid_alpha=0.5)
plt.grid()
plt.show()
```

Аргумент *axis* указывает ось ('x', 'y' или 'both' - 'обе'), к которой должны быть применены параметры. 
*direction* помещает галочки внутри осей, снаружи осей или и то, и другое. 
Аргументы *length* и *width* задают длину метки и ширину метки (в дробных значениях). 
*labelcolor* присваивает цвета меткам. 
Аргумент *colors* задает цвет галочки и цвет метки. 
*grid_color* задает цвет сетки, а *grid_alpha* задает непрозрачность сетки. 
plt.tick_params() можно использовать для изменения важных свойств сетки, таких как grid_color(аргумент - код цвета), grid_alpha(аргумент - дробное число), grid_linewidth(аргумент - дробное число) и grid_linestyle(аргумент - строка).

### Добавление легенды

Легенда - это формулировка на карте или диаграмме, объясняющая используемые на ней символы. 
Метод plt.legend() помещает легенду на оси.

Когда на графике присутствует несколько строк, код немного отличается от обычной практики.

```python
import numpy as np
import matplotlib.pyplot as plt
x = np.linspace(0, 7, 100)
line1, = plt.plot(x, np.sin(x), label='sin')
line2, = plt.plot(x, np.cos(x), label='cos')
plt.legend(handles=[line1, line2], loc='best')
plt.grid(b=True, color='aqua', alpha=0.6, linestyle='dashdot')
plt.show()
```

Положение условных обозначений также можно изменить, установив соответствующие значения (
'lower right' - "нижний правый", 
'lower left'  - "нижний левый", 
'upper right' - "верхний правый", 
'upper left'  - "верхний левый" и даже 
'best' - "наилучшим способом") в loc.

    plt.legend(handles=[line1, line2], loc='lower right')

Давайте добавим все эти функции в наш код и посмотрим, как выглядит график 1.

```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as stats
mu = 0
std = 1
x = np.linspace(start=-4, stop=4, num=100)
y = stats.norm.pdf(x, mu, std) 
plt.plot(x, y, label='PDF')
plt.xlabel('x', fontsize=15)
plt.ylabel('PDF', fontsize=15)
plt.grid(b=True, which='major', color='DarkTurquoise', alpha=0.4, linestyle=':', linewidth=2)
plt.minorticks_on()
plt.grid(b=True, which='minor', color='beige', alpha=0.2, linestyle='-', linewidth=2)
plt.legend()
plt.show()
```

### Увеличение размера фигуры

По умолчанию размер графика, отображаемого Jupyter notebook, довольно мал. 
Размер по умолчанию составляет всего 640x480. Однако сохраненные изображения имеют еще меньшие размеры.

```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as stats
x = np.linspace(start=-4, stop=4, num=100)
y = stats.norm.pdf(x, 0, 1) 
plt.plot(x, y)
plt.show()
```

Сохраненный файл имеет размеры 433px * 288px. 
Давайте изменим настройки, чтобы иметь пользовательский размер для графика. 
plt.rcParams[] предназначен для этой конкретной цели. Общий синтаксис выглядит следующим образом:

    plt.rcParams['figure.figsize'] = [width, height]

Matplotlib использует конфигурационные файлы matplotlibrc для настройки всех видов свойств, которые мы называем "rc-настройками" или "rc-параметрами’. Можно управлять значениями по умолчанию практически для каждого свойства в Matplotlib: размер рисунка и DPI, ширина линии, цвет и стиль, оси, свойства оси и сетки, свойства текста и шрифта и так далее. Как только файл matplotlibrc будет найден, он не будет искать никаких других путей. Местоположение текущего активного файла matplotlibrc можно увидеть, введя следующие команды:

    import matplotlib
    matplotlib.matplotlib_fname()

### Перевод в Сантиметры

Важным моментом, который следует отметить здесь, является то, что кортеж figsize принимает значения в дюймах. 
Чтобы перевести в сантиметр (см), разделите значение на 2,54, так как 1 дюйм = 2,54 см.

    plt.rcParams['figure.figsize'] = [10/2.54, 8/2.54]

Теперь попробуйем настроить размер Рисунка-2 с помощью rcParams.

```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as stats
mu = 0
std = 1
x = np.linspace(start=-4, stop=4, num=100)
y = stats.norm.pdf(x, mu, std) 
plt.plot(x, y, label='PDF')
plt.xlabel('x', fontsize=15)
plt.ylabel('PDF', fontsize=15)
plt.grid(b=True, color='DarkTurquoise', alpha=0.2, linestyle=':', linewidth=2)
plt.rcParams['figure.figsize'] = [15/2.54, 7/2.54]
plt.legend()
plt.show()
```

### Настройки стиля

plt.style.use() можно использовать для переключения на простые в использовании стили построения графиков. 
Пакет стилей предоставляет широкий спектр предустановленных стилей, что делает графики привлекательными, а их размеры увеличенными.

    plt.style.use('classic')

Команда plot.style.available выдаёт список всех различные стили, доступных для использования.

```python
print(plt.style.available)
```

Предположим, если мы хотим иметь темный фон для графика, то plt.style.use(‘dark_background’) будет служить этой цели.

```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as stats
plt.style.use('dark_background')
mu = 0
std = 1
x = np.linspace(start=-4, stop=4, num=100)
y = stats.norm.pdf(x, mu, std) 
plt.plot(x, y, label='PDF')
plt.xlabel('x', fontsize=15)
plt.ylabel('PDF', fontsize=15)
plt.grid(b=True, color='DarkTurquoise', alpha=0.2, linestyle=':', linewidth=2)
plt.legend()
plt.show()
```
